```python
from keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
model=load_model('shape.h5')
```

```python
from skimage.transform import resize
```

```python
def detect(frame):
    try:
        img=resize(frame,(64,64))
        img=np.expand_dims(img,axis=0) #we hav given 64,64,  3(rgb) 
        if(np.max(img)>1):    #dimention of matrix
            img=img/255.0
        prediction= model.predict(img)
        print(prediction)
        print(model.predict_classes(img))
    except AttributeError :
        print("Shape not found")
```

```python
img = cv2.imread('shape5.jpg')
plt.imshow(img)
```

```python
frame = cv2.imread(r'shape5.jpg')
data= detect(frame)
```
